```python
def IntelNet(input_shape = (21,1), classes = 2):
    # Define the input as a tensor with shape input_shape
    
    X_input = Input(input_shape)
   
    
    # Zero-Padding
    #X = ZeroPadding1D(padding=1)(X_input)
   
    
     # Stage 1
    X = Conv1D(128, 7, strides = 1, name = 'conv1',kernel_initializer='glorot_uniform', padding = 'SAME')(X_input)
    X = BatchNormalization(name = 'bn_conv1')(X)
    X = Activation('relu')(X)
    #X = MaxPooling1D((3), strides=(2),padding = 'SAME')(X)
    
   # print(X)
   
    X = Conv1D(50, 18, strides = 2, name = 'conv2')(X)
    X = BatchNormalization(name = 'bn_conv2')(X)
    X = Activation('relu')(X)
    #X = MaxPooling1D((3), strides=(2))(X)
    
    X = Conv1D(60, 19, strides = 2, name = 'conv3')(X)
    X = BatchNormalization(name = 'bn_conv3')(X)
    X = Activation('relu')(X)
    
    
    X = Conv1D(70, 20, strides = 2, name = 'conv4')(X)
    X = BatchNormalization(name = 'bn_conv4')(X)
    X = Activation('relu')(X)
  #  X = MaxPooling1D((3), strides=(2))(X)
    
    
    
    
    
    
  
    
    # output layer
    X = Flatten()(X)
    X = Dense(classes, activation='softmax')(X)
    
    
    # Create model
    model = Model(inputs = X_input, outputs = X, name='ResNet50')

    return model
```

```python
model = IntelNet(input_shape = (21,1), classes = 2)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```
